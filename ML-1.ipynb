## איחוד טבלאות לפי קבוצה כולל הוספת מיקום

```python
def data_prepare (team_general, team_stat, location):
    # הוספת מיקום בטבלה של קבוצה יריבה
    team_main = pd.merge(team_general, location, on='Opponent', how='left')
    
    # סידור עמודות משנה לשמות של עמודה רגילה בטבלת STAT
    new_columns = [col[1] for col in team_stat.columns]
    team_stat.columns = new_columns
    #print(team_stat.columns)
    
    # איחוד טבלאות לטבלה אחת
    team_main = pd.merge(team_main, team_stat, on='Round', how='outer')
    
    # הסרת העמודות הכפולות
    team_main = team_main.drop(columns=['Date_y', 'Time_y', 'Day_y', 'Venue_y', 'Result_y', 'GF_y', 'GA_y', 'Opponent_y','Match Report_y'])
    
    # הורדת תווים מיותרים שנוצרו מכפילות עמודות
    team_main.columns = [col.replace('_x', '') for col in team_main.columns]
    team_main['Round'] = team_main['Round'].str.replace('Matchweek ', '')
    
    # הוספת עמודה רק עם החודש של המחשק מתוך התארריך
    #team_main['Month'] = team_main['Date'].astype(int).astype(str)
    # Assuming 'df' is your DataFrame
    #df['Month'] = df['Date'].str.split('-', expand=True)[1]
    team_main['Date'] = pd.to_datetime(team_main['Date'])
    team_main['month'] = team_main['Date'].dt.month
    team_main['year'] = team_main['Date'].dt.year
    

    # הסרת עמודות לא רלוונטיות או כפולות במידע
    team_main = team_main.drop(columns=['Attendance','Notes','Won%', 'Match Report', 'Opponent','Date'])
    
    # הסרת שורה אחרונה
    team_main = team_main.drop(index=38)
    
    return team_main
```

```python
# ארסנל
arsenal19_20 = data_prepare(as_sf_19_20, as_ms_19_20, loc_19_20 )
arsenal20_21 = data_prepare(as_sf_20_21, as_ms_20_21, loc_20_21)
arsenal21_22 = data_prepare(as_sf_21_22, as_ms_21_22, loc_21_22 )
arsenal22_23 = data_prepare(as_sf_22_23, as_ms_22_23, loc_22_23 )

# ליברפול
liverpool19_20 = data_prepare(lv_sf_19_20, lv_ms_19_20, loc_19_20 )
liverpool20_21 = data_prepare(lv_sf_20_21, lv_ms_20_21, loc_20_21 )
liverpool21_22 = data_prepare(lv_sf_21_22, lv_ms_21_22, loc_21_22 )
liverpool22_23 = data_prepare(lv_sf_22_23, lv_ms_22_23, loc_22_23 )
```

# PCA 

### כדי להחליט לכמה מימדים אנחנו רוצים לצמצם את הדאטה נחשב את יחס השונות המוסבר המצטבר

```python
# Drop rows with missing values
normal_groups_arsenal.dropna(inplace=True)
```

```python
# Create a PCA instance
pca = PCA()
pca.fit(normal_groups_arsenal)

# Compute the cumulative explained variance ratio
cumulative_var_ratio = np.cumsum(pca.explained_variance_ratio_)
```

### יחס השונות המוסבר המצטבר   

```python
plt.plot(range(1, len(cumulative_var_ratio) + 1), cumulative_var_ratio)
plt.xlabel('Number of Components',fontsize=15)
plt.ylabel('Cumulative Explained Variance Ratio')
plt.title('Cumulative Explained Variance Ratio',fontsize=17)

plt.axvline(x=20, color='red', linestyle='--')

plt.show()
```

# K-mean  אלגוריתם מבוסס מרחק 

```python
df_pca.columns = df_pca.columns.astype(str)
```

```python
distortions = []
K = range(1, 11)  # Test for different values of k (e.g., 1 to 10)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_pca)
    distortions.append(kmeans.inertia_)

# Plot the elbow curve
plt.plot(K, distortions, 'bx-')
plt.xlabel('Number of Clusters (k)', fontsize=15)
plt.ylabel('Distortion')
plt.title('Elbow Method')

plt.axvline(x=3, color='red', linestyle='--')

plt.show()
```

#### ככל שנגדיל את מספר הקבוצות שאנו רוצים לסווג אליהם השונות תקטן. לפי הגרף נבחר לסווג לפי 3 קבוצות כי שם השיפוע מתחיל להיות פחות חד כלומר כל הגדלה של מספר הקבוצות (K) הוא פחות משמעותי   

```python
data = df_pca

#Choose the number of clusters
k = 3

# Initialize the k-means algorithm
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)


# Fit the algorithm to the data
kmeans.fit(data)

# Get the cluster labels for each data point

cluster_labels = kmeans.labels_
# Add the cluster labels back to the dataframe
df_pca['Cluster'] = cluster_labels
#df_pca
```

## גרף 1 - לכל אשכול את הגולים שהבקיעו והגולים שקבוצה יריבה הבקיעה

```python
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Generate sample data using make_blobs
X, _ = make_blobs(n_samples=150, centers=3, cluster_std=0.60, random_state=0)

# Perform hierarchical clustering
agglomerative_clustering = AgglomerativeClustering(n_clusters=3)
clusters = agglomerative_clustering.fit_predict(X)

# Assuming you have a DataFrame 'df' containing the data
# Replace the values below with your actual data
import pandas as pd
df = pd.DataFrame(X, columns=['GA', 'GF'])
df['Cluster'] = clusters

# Visualize clusters using scatter plot
plt.figure(figsize=(8, 6))

# Plot each cluster with a different color
for cluster in df['Cluster'].unique():
    cluster_data = df[df['Cluster'] == cluster]
    plt.scatter(cluster_data['GA'], cluster_data['GF'], label=f'Cluster {cluster}')

plt.xlabel('Goals Scored')
plt.ylabel('Goals Conceded')
plt.title('Cluster Visualization')
plt.legend()
plt.grid(True)
plt.show()
```

### נבחר לכמה קבוצות לסווג את הדאטה 

```python
distortions = []
K = range(1, 11)  # Test for different values of k (e.g., 1 to 10)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42,n_init = 10)
    kmeans.fit(df_pca_lv)
    distortions.append(kmeans.inertia_)

# Plot the elbow curve
plt.plot(K, distortions, 'bx-')
plt.xlabel('Number of Clusters (k)', fontsize=15)
plt.ylabel('Distortion')
plt.title('Elbow Method')

plt.axvline(x=4, color='red', linestyle='--')

plt.show()
```

#### ככל שנגדיל את מספר הקבוצות שאנו רוצים לסווג אליהם השונות תקטן. לפי הגרף נבחר לסווג לפי 4 קבוצות כי שם השיפוע מתחיל להיות פחות חד כלומר כל הגדלה של מספר הקבוצות (K) הוא פחות משמעותי   

```python
data = df_pca_lv

# Choose the number of clusters
k = 4

# Initialize the k-means algorithm
kmeans = KMeans(n_clusters=k, random_state=42,n_init=10)

# Fit the algorithm to the data
kmeans.fit(data)

# Get the cluster labels for each data point
cluster_labels = kmeans.labels_

# Add the cluster labels back to the dataframe
df_pca_lv['Cluster'] = cluster_labels
df_pca_lv
```
